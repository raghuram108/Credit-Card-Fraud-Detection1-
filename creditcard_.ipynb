Under Sampling

Build a sample dataset containing similar distribution of normal transactons and fraudulent transactions


Number of Fraudulent transactions --> 492

```python
legit_sample = legit.sample(n=25)
```

concatenating two dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0 )
```

```python
new_dataset.head(5)
```

```python
new_dataset.tail(5)
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the dataset into Features & Targets

```python
X = new_dataset.drop(columns = 'Class', axis =1)
Y = new_dataset['Class']
```

```python
print(Y)
```

Split the data into Training data and testing data

All the features of training data distroned in X_train,
all the lables for all those coresponding data will store in X_test

```python
X_train, X_test, Y_train,Y_test = train_test_split(X,Y, test_size=0.5, stratify= Y, random_state = 2 )
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Trainig

Logistic Regression - for bineary classifcation programes

```python
model = LogisticRegression()
```

```python
 # training the Logistic Regression Model with Traning Data
```

```python
model.fit(X_train, Y_train)
```

Model Evalution::::::

Accuracy Score

```python
# accuracy on  training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy )
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy )
```
